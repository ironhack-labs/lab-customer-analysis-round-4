```python
#importing libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
#reading file
mkt = pd.read_csv(r"C:\Users\pedro\Desktop\Ironhack\lab-customer-analysis-round-4\files_for_lab\csv_files\marketing_customer_analysis.csv")
mkt.head(3)
```

```python
#standardization of columns names

cols = []
for i in range(len(mkt.columns)):
    cols.append(mkt.columns[i].lower().replace(' ','_'))

mkt.columns=cols    
list(mkt.columns)
```

```python
#checking variable types
mkt.dtypes
```

```python
#numeric data into dataframe
numerical = mkt.select_dtypes(include=[np.number])
numerical.dtypes
```

```python
#categorical data into dataframe
categoricals = mkt.select_dtypes(['object'])
categoricals.dtypes
```

```python
#using seaborn to construct histograms
#none of the distributions looks like a normal distribution

fig, axes = plt.subplots(2, 4, figsize=(18,6))

i=0
for j in range(2):
    for k in range(4):
        sns.distplot(numerical.iloc[:,[i]],ax=axes[j,k])
        axes[j,k].set_title(list(numerical.columns.values)[i])
        i=i+1
        
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.8)
plt.show()
```

```python
#using matplotlib to construct histograms

fig, axes = plt.subplots(2, 4, figsize=(18,6))

i=0
for j in range(2):
    for k in range(4):
        axes[j,k].hist(numerical.iloc[:,[i]])
        axes[j,k].set_title(list(numerical.columns.values)[i])
        i=i+1
        
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.8)
plt.show()
```

```python
correlations_matrix = numerical.corr()
correlations_matrix
```

```python
#checking correlations between numerical variables
sns.heatmap(correlations_matrix, annot=True)
plt.show()
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```

```python
#no correlations above 0.9 so there is no multicolinearity 
#checking which variables are correlated above 0.3
    
correlations_matrix = numerical.corr()
correlated_variables = correlations_matrix[abs(correlations_matrix) >= 0.3]
correlated_variables = correlated_variables.mask(np.tril(np.ones(correlated_variables.shape)).astype(bool)) 
correlated_variables_pairs = [(col1,col2) for col1 in correlated_variables.columns for col2 in correlated_variables.columns if not pd.isna(correlated_variables[col1][col2])]
print("variables correlated above 0.3 are: \n\n",correlated_variables_pairs)
print("\n\n variables correlated above 0.3 with 'total claim amount' are: \n\n", pd.DataFrame([k for k in correlated_variables_pairs if 'total_claim_amount' in k])[1])
    
```
