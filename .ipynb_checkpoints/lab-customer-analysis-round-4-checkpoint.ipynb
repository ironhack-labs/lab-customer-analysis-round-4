```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
root = r'C:\Users\aidag\OneDrive\Escritorio\ironhack\lab-customer-analysis-round-4\files_for_lab\csv_files\marketing_customer_analysis.csv'
original_csv = pd.read_csv(root)
original_csv.head(5)
```

```python
original_csv.dtypes
```

```python
columns_to_convert_to_categorical = ["Customer", "State", "Response", "Coverage", "Education", "Effective To Date", "EmploymentStatus", "Gender", "Location Code", "Marital Status", "Policy Type", "Policy", "Sales Channel", "Vehicle Class", "Vehicle Size"]

for column in columns_to_convert_to_categorical:
    original_csv[column] = original_csv[column].astype("category")

numerical_data = original_csv.select_dtypes("number")
categorical_data = original_csv.select_dtypes("category")
```

```python
display(numerical_data)
```

```python
display(categorical_data)
```

```python
for col in numerical_data.columns:
    print(col)
    sns.histplot(numerical_data[col])
    plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=len(numerical_data.columns), figsize=(25, 5))

for i, col in enumerate(numerical_data.columns):
    axs[i].hist(numerical_data[col], bins=10, edgecolor='black', color='skyblue')
    axs[i].set_title(col)
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
numerical_data = original_csv.select_dtypes('number')
correlation_matrix = numerical_data.corr()
display(correlation_matrix)
```

```python

mask = np.zeros_like(correlation_matrix)
mask[np.triu_indices_from(mask)] = True 

fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlation_matrix, mask=mask, annot=True)
plt.show()
```
