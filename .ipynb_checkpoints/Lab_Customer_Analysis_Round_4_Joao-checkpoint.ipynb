```python
# Importing libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing file

df = pd.read_excel(r"C:\Users\joaoa\Desktop\Ironhack\Labs\lab-customer-analysis-round-4\files_for_lab\excel_files\marketing_customer_analysis.xlsx")
df
```

```python
# Standardizing columns names

cols = []
for i in range(len(df.columns)):
    cols.append(df.columns[i].lower().replace(" ","_"))
df.columns = cols
df
```

```python
# Looking for data types

df.dtypes
```

```python
# Creating numerical dataframe

numerical = df.select_dtypes(include=np.number)
numerical
```

```python
# Creating categorical dataframe

categoricals = df.select_dtypes(include = object)
categoricals
```

```python
# Using seaborn library to construct distribution plots

sns.distplot(numerical["customer_lifetime_value"])
plt.xlabel("Customer lifetime value")
plt.ylabel("Density")
plt.title("'Customer lifetime value' density plot")
plt.show()

sns.distplot(numerical["income"])
plt.xlabel("Income")
plt.ylabel("Density")
plt.title("'Income' density plot")
plt.show()

sns.distplot(numerical["monthly_premium_auto"])
plt.xlabel("Monthly premium auto")
plt.ylabel("Density")
plt.title("'Monthly premium auto' density plot")
plt.show()

sns.distplot(numerical["months_since_last_claim"])
plt.xlabel("Months since last claim")
plt.ylabel("Density")
plt.title("'Months since last claim' density plot")
plt.show()

sns.distplot(numerical["months_since_policy_inception"])
plt.xlabel("Months since policy inception")
plt.ylabel("Density")
plt.title("'Months since policy inception' density plot")
plt.show()

sns.distplot(numerical["number_of_open_complaints"])
plt.xlabel("Number of open complaints")
plt.ylabel("Density")
plt.title("'Number of open complaints' density plot")
plt.show()

sns.distplot(numerical["number_of_policies"])
plt.xlabel("Number of policies")
plt.ylabel("Density")
plt.title("'Number of policies' density plot")
plt.show()

sns.distplot(numerical["total_claim_amount"])
plt.xlabel("Total claim amount")
plt.ylabel("Density")
plt.title("'Total claim amount' density plot")
plt.show()
```

```python
# Using Matplotlib to contruct histograms

plt.hist(numerical["customer_lifetime_value"])
plt.xlabel("Customer lifetime value")
plt.ylabel("Quantity")
plt.title("'Customer lifetime value' histogram")
plt.show()

plt.hist(numerical["income"])
plt.xlabel("Income")
plt.ylabel("Quantity")
plt.title("'Income' histogram")
plt.show()

plt.hist(numerical["monthly_premium_auto"])
plt.xlabel("Monthly premium auto")
plt.ylabel("Quantity")
plt.title("'Monthly premium auto' histogram")
plt.show()

plt.hist(numerical["months_since_last_claim"])
plt.xlabel("Months since last claim")
plt.ylabel("Quantity")
plt.title("'Months since last claim' histogram")
plt.show()

plt.hist(numerical["months_since_policy_inception"])
plt.xlabel("Months since policy inception")
plt.ylabel("Quantity")
plt.title("'Months since policy inception' histogram")
plt.show()

plt.hist(numerical["number_of_open_complaints"])
plt.xlabel("Number of open complaints")
plt.ylabel("Quantity")
plt.title("'Number of open complaints' histogram")
plt.show()

plt.hist(numerical["number_of_policies"])
plt.xlabel("Number of policies")
plt.ylabel("Quantity")
plt.title("'Number of policies' histogram")
plt.show()

plt.hist(numerical["total_claim_amount"])
plt.xlabel("Total claim amount")
plt.ylabel("Quantity")
plt.title("'Total claim amount' histogram")
plt.show()
```

```python
# Creating a new dataframe showing the correlations

correlations_matrix = numerical.corr()
correlations_matrix
```

```python
# Checking the multicollinearity between the features

mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True
ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlations_matrix, mask = mask, annot = True)
plt.show()
```
