```python
# creating bar PLot for all the data



def histplot(data,columns):
    
    for column in columns:
       sns.histplot(x =data[column],bins='auto')
       plt.xlabel(column)
       plt.ylabel('Value')
       plt.title(f'barplot for  {column}')
       plt.show()

data = numericals
columns = ['customer_lifetime_value', 'income', 'monthly_premium_auto',
       'months_since_last_claim', 'months_since_policy_inception',
       'number_of_open_complaints', 'number_of_policies','total_claim_amount'
       ]

histplot(data,columns)




# Do the distributions for different numerical variables look like a normal distribution
#customer_lifetime_value', NO 
#  'income', Not Really
# 'monthly_premium_auto',  NO 
#'months_since_last_claim',  NO 
# 'months_since_policy_inception', Look like the variable look like  a normal distribution
# number_of_open_complaints', NO
# 'number_of_policies',NO 
# total_claim_amount'NO looks like a positive relation between variable



```
