```python
def histograms (dataframe):
    

    for column in numerical_data:
        
        plt.hist(numerical_data[column], bins=40)

        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title('Normalization')

        plt.show()
    return 

histograms (numerical_data)

#only the total_claim amount fallows a normalized distribution
```

```python
#Checking the multicollinearity
correlation_matrix = numerical_data.corr()
correlation_matrix
```

```python
#make the sns heatmap and correlation map

sns.heatmap(correlation_matrix, annot=True)
plt.show()
mask = np.zeros_like(correlation_matrix)
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlation_matrix, mask=mask, annot=True)
plt.show()

#there is no correlations above 0.9, so no variable will be droped.
```
