```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
mc = pd.read_csv('marketing_customer_analysis.csv')
mc
```

```python
# Check the data types of the columns
```

```python
mc.dtypes
```

```python
cols = []
for i in range(len(mc.columns)): 
    cols.append(mc.columns[i].lower().replace(' ', '_'))
mc.columns = cols
mc
```

```python
# Get the numeric data into dataframe called numerical and categorical columns in a dataframe called categoricals.
```

```python
mc_c = mc.select_dtypes(include = 'object')
mc_c
```

```python
mc_n = mc.select_dtypes(include = 'number')
mc_n
```

```python
# Use seaborn library to construct distribution plots for the numerical variables
```

```python
for i in mc_n.columns:
    sns.distplot(mc_n[i], kde = True)
    plt.show()
```

```python
# Use Matplotlib library to construct distribution plots for the numerical variables
```

```python
for i in mc_n.columns:
    plt.hist(mc_n[i], orientation='horizontal')
    plt.xlabel(i)
    plt.show()
```

```python
# Do the distributions for different numerical variables look like a normal distribution
```

```python
# For the numerical variables, check the multicollinearity between the features. 
# Please note that we will use the column total_claim_amount later as the target variable.
```

```python
mc_n.corr()

# there is no multicollianearity. 
```
